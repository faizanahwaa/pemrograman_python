```
import numpy as np
import matplotlib.pyplot as plt
```

```
frek = [700, 900, 1800]
tipe_pilihan = ['urban' , 'suburban' , 'open']
ht = 80 #meter
hr = 1.5 #meter
d = np.arange(200,700)
#print(d)
```

```
#FUNGSI PATHLOSS
def cos(frek,tipe,ht,hr,d):
  #frek = frekuensi (MHz)
  #tipe = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
#FUNGSI GRAFIK ADA 3 TIPE BERBEDA TETAPI PADA FREKUENSI 1800 MHz
 
dict_g1 = {}
 
def grafik1_1800(frek,tipe_pilihan):
  f1 = frek[2]
  for i in range(len(tipe_pilihan)):
    tipe = tipe_pilihan[i]
    #print(tipe)
    for i in range(len(d)):
      PL = cos(f1,tipe,ht,hr,d)
    dict_g1[tipe]= PL
 
  #print(PL)
  print(dict_g1)
  #print(dict_g1['urban'])
grafik1_1800(frek,tipe_pilihan)
```

```
#FUNGSI GRAFIK ADA 3 FREKUENSI BERBEDA TAPI TIPE URBAN
 
dict_g2 = {}
 
def grafik2_urban(frek,tipe_pilihan):
  tipe = tipe_pilihan[0]
  for i in range(len(frek)):
    f2 = frek[i]
    #print(tipe)
    for i in range(len(d)):
      PL = cos(f2,tipe,ht,hr,d)
    dict_g2[f2]= PL
 
  #print(PL)
  print(dict_g2)
  #print(dict_g1['urban'])
grafik2_urban(frek,tipe_pilihan)
```

```
#PLOT GRAFIK 1
def graph_1():
  plt.figure()
  plt.plot(d,dict_g1['urban'],'b', label='Urban')
  plt.plot(d,dict_g1['open'],'r', label='Open')
  plt.plot(d,dict_g1['suburban'],'g', label='Suburban')
  plt.legend()
  plt.title("Frekuensi 1800 MHz")
  plt.xlabel("Jarak (m)")
  plt.ylabel("Path Loss (dB)")
  plt.show()
 
#PLOT GRAFIK 2
def graph_2():
  plt.figure()
  plt.plot(d,dict_g2[700],'b', label='700 MHz')
  plt.plot(d,dict_g2[900],'r', label='900 MHz')
  plt.plot(d,dict_g2[1800],'g', label='1800 MHz')
  plt.legend()
  plt.title("Model Urban")
  plt.xlabel("Jarak (m)")
  plt.ylabel("Path Loss (dB)")
  plt.show()
```
